```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

response = requests.get('https://www.vienna.at/features/bezirke-wien')
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
# Convert to string
pretty = soup.prettify()
#print(pretty)
```

```python
bezirkes = soup.select('ul[id=bezirke] li', limit = 23)    # Soup object type, total 23 bezirkes in Vienna
ordernumbers = []
names        = []
addresses    = []
```

```python
# Go through all bezirkes in the list
for bezirke in bezirkes:

    # Get bezirke order number, name and address
    bezirkeData = bezirke.text.split('\n')
    ordernumbers.append(bezirkeData[1])
    names.append(bezirkeData[2])
    addresses.append(bezirkeData[3])
    
```

```python
print(ordernumbers, names, addresses)
```

```python
# Create a pandas dataframe
df = pd.DataFrame({'BezirkeNbr':ordernumbers, 'Bezirke':names, 'Zip':addresses})
df.head(23)
```

```python
df.to_csv('bezirkes.csv', index=False, encoding='utf8')
```
